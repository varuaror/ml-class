```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1" #Do not use GPU from TF
import numpy as np

from skimage import io
from skimage import data
from scipy.signal import convolve2d
```

```python
image = io.imread('dog.jpg', as_gray=True)
#3x3 convolution, a matrix
#the network basically learns which kernel numbers generate the most meaningful features -- becomes the weights
#retains spacial information but detecting things with the feature maps
kernel = [[1.0, -0.25, 0.0],
          [-0.25, 2.0, -0.25],
          [0.0, -0.25, -1.0]]
new_image = convolve2d(image, kernel).clip(0.0, 1.0)
io.imshow(new_image)
```
